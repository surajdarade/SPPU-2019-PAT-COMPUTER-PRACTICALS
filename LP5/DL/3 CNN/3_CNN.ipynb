```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
train_df = pd.read_csv('fashion-mnist_train.csv')
test_df = pd.read_csv('fashion-mnist_test.csv')
```

```python
class_names = ['T-shirt/top', 'Trouser', 'Pullover', 'Dress', 'Coat','Sandal', 'Shirt', 'Sneaker', 'Bag', 'Ankle boot']
```

```python
x_train = train_df.iloc[:,1:].to_numpy()
x_train = x_train.reshape([-1,28,28,1])
x_train = x_train / 255
```

```python
y_train = train_df.iloc[:,0].to_numpy()
```

```python
x_test = test_df.iloc[:,1:].to_numpy()
x_test = x_test.reshape([-1,28,28,1])
x_test = x_test / 255
```

```python
y_test = test_df.iloc[:,0].to_numpy()
```

```python
plt.figure(figsize=(10,10))
for i in range(25):
    plt.subplot(5,5,i+1)
    plt.xticks([])
    plt.yticks([])
    plt.imshow(x_train[i], cmap=plt.cm.binary)
    plt.xlabel(class_names[y_train[i]])
plt.show()
```

```python
import keras
from keras.models import Sequential
from keras.layers import Dense,Conv2D,Flatten,MaxPooling2D,Dropout

model=keras. Sequential([keras.layers.Input(shape=(28, 28, 1)),

keras.layers.Conv2D(64, (3,3), activation='relu', padding='same'),
keras.layers.BatchNormalization(),
keras.layers.Conv2D(64, (3,3), activation='relu', padding='same'),
keras.layers.BatchNormalization(),
keras.layers. MaxPooling2D((2,2)),
keras.layers.Dropout(0.3),

keras.layers.Conv2D(128, (3,3), activation='relu', padding='same'),
keras.layers.BatchNormalization(),
keras.layers.Conv2D(128, (3,3), activation='relu', padding='same'),
keras.layers.BatchNormalization(),
keras.layers.MaxPooling2D((2,2)),
keras.layers. Dropout (0.4),

keras.layers.Conv2D(256, (3,3),  activation='relu', padding='same'),
keras.layers.BatchNormalization(),
keras.layers.Conv2D(256, (3,3),  activation='relu', padding='same'),
keras.layers.BatchNormalization(),
keras.layers.MaxPooling2D((2,2)),
keras.layers. Dropout (0.5),

keras.layers.Flatten(),
keras.layers. Dense (512, activation='relu'),
keras.layers.BatchNormalization(),
keras.layers. Dropout(0.5),
keras.layers. Dense(10, activation='softmax')
])
```

```python
model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',  # or another depending on your task
    metrics=['accuracy']
)
model.fit(x_train,y_train,epochs=50,batch_size=1200,validation_split=0.05)

```

```python
output = model.evaluate(x_test,y_test)
```

```python
print(f"Loss: {output[0]*100}%"
      ,f"Accuracy: {output[1]*100}%",sep="\n")
```

```python
y_probas = model.predict(x_test)
y_pred = y_probas.argmax(axis=-1)
y_pred
```

```python
plt.figure(figsize=(10,10),)
for i in range(25):
    plt.subplot(5,5,i+1)
    plt.xticks([])
    plt.yticks([])
    plt.grid(False)
    plt.imshow(x_test[i], cmap=plt.cm.binary)
    plt.title(f"Pred:{class_names[y_pred[i]]}")
plt.show()
```

```python
from sklearn.metrics import classification_report

num_classes = 10
class_names = ["class {}".format(i) for i in range(num_classes)]
cr = classification_report(y_test, y_pred, target_names=class_names)
print(cr)
```
