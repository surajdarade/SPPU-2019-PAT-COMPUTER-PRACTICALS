```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd 
```

```python
df = pd.read_csv("Iris.csv")
df.head()
```

```python
plt.figure(figsize=(10, 8))
plt.subplot(2, 2, 1)
plt.hist(df['SepalLengthCm'], bins=20, color='skyblue', edgecolor='black')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Frequency')
plt.title('Histogram of Sepal Length')
```

```python
plt.subplot(2, 2, 2)
plt.hist(df['SepalWidthCm'], bins=20, color='salmon', edgecolor='black')
plt.xlabel('Sepal Width (cm)')
plt.ylabel('Frequency')
plt.title('Histogram of Sepal Width')
```

```python
plt.subplot(2, 2, 3)
plt.hist(df['PetalLengthCm'], bins=20, color='lightgreen', edgecolor='black')
plt.xlabel('Petal Length (cm)')
plt.ylabel('Frequency')
plt.title('Histogram of Petal Length')
```

```python
plt.subplot(2, 2, 4)
plt.hist(df['PetalWidthCm'], bins=20, color='lightcoral', edgecolor='black')
plt.xlabel('Petal Width (cm)')
plt.ylabel('Frequency')
plt.title('Histogram of Petal Width')
```

```python
plt.figure(figsize=(10, 8))

# Sepal Length
plt.subplot(2, 2, 1)
plt.boxplot(df['SepalLengthCm'])
plt.xlabel('Sepal Length (cm)')
plt.title('Boxplot of Sepal Length')

# Sepal Width
plt.subplot(2, 2, 2)
plt.boxplot(df['SepalWidthCm'])
plt.xlabel('Sepal Width (cm)')
plt.title('Boxplot of Sepal Width')

# Petal Length
plt.subplot(2, 2, 3)
plt.boxplot(df['PetalLengthCm'])
plt.xlabel('Petal Length (cm)')
plt.title('Boxplot of Petal Length')

# Petal Width
plt.subplot(2, 2, 4)
plt.boxplot(df['PetalWidthCm'])
plt.xlabel('Petal Width (cm)')
plt.title('Boxplot of Petal Width')

plt.tight_layout()
plt.show()
```
