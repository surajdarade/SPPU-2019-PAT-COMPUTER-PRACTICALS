```python
df = df.rename(columns={'Iris-setosa': 'Species'})
```

```python
df.head(5)
```

```python
# Filter the DataFrame to include only the specified species
setosa_stats = df[df['Species'] == 'Iris-setosa'].describe()
versicolor_stats = df[df['Species'] == 'Iris-versicolor'].describe()
virginica_stats = df[df['Species'] == 'Iris-virginica'].describe()

# Display basic statistical details for each species
print("Statistical details for Iris-setosa:")
print(setosa_stats)
print()

print("Statistical details for Iris-versicolor:")
print(versicolor_stats)
print()

print("Statistical details for Iris-virginica:")
print(virginica_stats)
```
